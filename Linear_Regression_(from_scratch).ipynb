```
import numpy as np
import random
from matplotlib import pyplot as plt
```

# [Plotting the results]

# 1.input data(2pt)

```
# data  m=100 
# a=2, b=5
m=100
a=2
b=5
sigma=0.1
x=np.array([random.random() for i in range(m)])
y=np.array([a*x_i+b+random.gauss(0,sigma) for x_i in x])
```

# 1. input data


```
plt.plot(x,[a*x_i+b for x_i in x],'blue',label='straight line')
plt.scatter(x,y,c='black',label='dot')
plt.legend(loc='best')
plt.show()
```

```
def linear_model(x,theta_0,theta_1):
    hypothesis=theta_0+theta_1*x
    return hypothesis

def objective_function(x,y,theta_0,theta_1):
    # mean squared error
    return ((linear_model(x,theta_0,theta_1)-y)**2).mean()/2

def gradient_descent(theta_0,theta_1,learning_rate):
    theta_0_=theta_0-learning_rate*((linear_model(x,theta_0,theta_1)-y).sum())
    theta_1_=theta_1-learning_rate*(((linear_model(x,theta_0,theta_1)-y)*x).sum())
    return theta_0_,theta_1_
```

```
# theta_0  1, theta_1  6
# learning rate 0.01
# until convergence
cost=[]

theta_0=1
theta_1=6
Theta_0=[theta_0]
Theta_1=[theta_1]
learning_rate=0.01
while True:
    if objective_function(x,y,theta_0,theta_1)<=6*1e-3:
        break
    
    cost.append(objective_function(x,y,theta_0,theta_1))
    theta_0,theta_1=gradient_descent(theta_0,theta_1,learning_rate)
    Theta_0.append(theta_0)
    Theta_1.append(theta_1)
```

# 2. Output results


```
plt.plot(x,theta_0+x*theta_1,'red',label='linear regression')
plt.scatter(x,y,c='black',label='dot')
plt.legend(loc='best')
plt.show()
```

# 3. plotting the energy values

```
#   6*1e-3   
plt.plot(cost)
```

# 4. Plotting the model parameters


```
plt.plot(range(len(Theta_0)),Theta_0,'red',label='theta0')
plt.legend(loc='best')
plt.show()
plt.plot(range(len(Theta_1)),Theta_1,'blue',label='theta1')
plt.legend(loc='best')
plt.show()
```
